```python
# Adaptation of the publicly available code of the NeurIPS 2020 paper entitled "TIM: Transductive Information Maximization":
# https://github.com/mboudiaf/TIM
import torch.nn.functional as F
from src.utils import get_one_hot, Logger, extract_features
from tqdm import tqdm
import torch
import time
from scipy.optimize import linear_sum_assignment
from sklearn.metrics import accuracy_score, f1_score
import numpy as np
from copy import deepcopy

import math

class KM(object):
    def __init__(self):
        self.device = "cpu"
        self.iter = 100
        self.alpha = 0
        self.eps = 1
        self.s_min = 0.001
        self.s_max = 100000
        self.init_info_lists()
        self.n_ways = n_ways

    def init_info_lists(self):
        self.timestamps = []
        self.criterions = []
        self.test_acc = []
        self.losses = []
        self.test_F1 = []

    def get_logits(self, samples, s):
        """
        inputs:
            samples : torch.Tensor of shape [n_task, shot, feature_dim]

        returns :
            logits : torch.Tensor of shape [n_task, shot, num_class]
        """
        n_tasks = samples.size(0)
#         logits = (- 2 * samples.matmul((self.s * self.weights).transpose(1, 2)) \
#                               + ((self.s * self.weights)**2).sum(2).view(n_tasks, 1, -1) \
#                               + ((self.s.unsqueeze(1)*samples.unsqueeze(2))**2).sum(3))  #
        w = self.weights  # N x K x C
        z = samples  # N x n x C
        
        diff = w.unsqueeze(1) - z.unsqueeze(2)  # N x n x K x C
        logits = ((diff ** 2) * s.unsqueeze(1)).sum(dim=-1)
        
        return logits  # N x n x K

    def A(self, p):
        """
        inputs:

            p : torch.tensor of shape [n_tasks, q_shot, num_class]
                where p[i,j,k] = probability of point j in task i belonging to class k
                (according to our L2 classifier)
        returns:
            v : torch.Tensor of shape [n_task, q_shot, num_class]
        """
        q_shot = p.size(1)
        v = p.sum(1) / q_shot
        return v

    def A_adj(self, v, q_shot):
        """
        inputs:
            V : torch.tensor of shape [n_tasks, num_class]
            q_shot : int
        returns:
            p : torch.Tensor of shape [n_task, q_shot, num_class]
        """
        p = v.unsqueeze(1).repeat(1, q_shot, 1) / q_shot
        return p
    
    def proj_box(self, x):
        """
        inputs:
            X : torch.tensor of shape [n_tasks, num_class, feature_dim, feature_dim]
        """
        args_min = 1 * (x >= self.s_min)
        args_max = 1 * (x < self.s_max)
        return x * args_min * args_max  + self.s_max * (1 - args_max) + self.s_min * (1 - args_min)
    
    def proj_trace(self, X):

        if self.diagonal == True:
            feature_dim = X.size(-1)
            aX = -X.sum(-1)
            b = - self.eps
            u = (aX - b).unsqueeze(-1)
            cond = (u <= 0)

            if self.unique_covariance == True:
                I = torch.ones(feature_dim).unsqueeze(0).to(self.device)
                return X * cond + (X + u / feature_dim * I ) * (1 - 1*cond)
            else:
                I = torch.ones(feature_dim).unsqueeze(0).unsqueeze(0).to(self.device)
                return X * cond + (X + u / feature_dim * I ) * (1 - 1*cond)
            
        else:
            feature_dim = X.size(-1)
            aX = -X.diagonal(offset=0, dim1=-1, dim2=-2).sum(-1)
            b = - self.eps
            u = (aX - b).unsqueeze(-1).unsqueeze(-1)
            cond = (u <= 0)

            if self.unique_covariance == True:
                I = torch.eye(feature_dim).unsqueeze(0).to(self.device)
                return X * cond + (X + u / feature_dim * I ) * (1 - 1*cond)
            else:
                I = torch.eye(feature_dim).unsqueeze(0).unsqueeze(0).to(self.device)
                return X * cond + (X + u / feature_dim * I ) * (1 - 1*cond)
        

    def get_preds(self, p):
        """
        inputs:
            p : torch.Tensor of shape [n_task, s_shot, feature_dim]

        returns :
            preds : torch.Tensor of shape [n_task, shot]
        """
        preds = p.argmax(2)
        return preds

    def init_weights(self, support, query, y_s, y_q):
        """
        inputs:
            support : torch.Tensor of shape [n_task, s_shot, feature_dim]
            query : torch.Tensor of shape [n_task, q_shot, feature_dim]
            y_s : torch.Tensor of shape [n_task, s_shot]
            y_q : torch.Tensor of shape [n_task, q_shot]

        updates :
            self.weights : torch.Tensor of shape [n_task, num_class, feature_dim]
        """
        t0 = time.time()
        n_tasks = support.size(0)
        one_hot = get_one_hot(y_s).float()
        counts = one_hot.sum(1).view(n_tasks, -1, 1)
        weights = one_hot.transpose(1, 2).matmul(support)
        self.weights = weights / counts

    def record_convergence(self, new_time, criterions):
        """
        inputs:
            support : torch.Tensor of shape [n_task, s_shot, feature_dim]
            query : torch.Tensor of shape [n_task, q_shot, feature_dim]
            y_s : torch.Tensor of shape [n_task, s_shot]
            y_q : torch.Tensor of shape [n_task, q_shot] :
        """
        self.criterions.append(criterions)
        self.timestamps.append(new_time)

    
    def record_info(self, y_q):
        """
        inputs:
            support : torch.Tensor of shape [n_task, s_shot, feature_dim]
            query : torch.Tensor of shape [n_task, q_shot, feature_dim]
            y_s : torch.Tensor of shape [n_task, s_shot]
            y_q : torch.Tensor of shape [n_task, q_shot] :
        """
        preds_q = self.p.argmax(2)
        n_tasks, q_shot = preds_q.size()
        accuracy = (preds_q == y_q).float().mean(1, keepdim=True)
        self.test_acc.append(accuracy)
        print("acc ", accuracy.mean())

        union = list(range(self.n_ways))
        for i in range(n_tasks):
            ground_truth = list(y_q[i].reshape(q_shot).cpu().numpy())
            preds = list(preds_q[i].reshape(q_shot).cpu().numpy())
            f1 = f1_score(ground_truth, preds, average='weighted', labels=union, zero_division=1)
            self.test_F1.append(f1)

    def get_logs(self):
        self.test_F1 = np.array([self.test_F1])
        self.criterions = torch.stack(self.criterions, dim=0).cpu().numpy()
        self.test_acc = torch.cat(self.test_acc, dim=1).cpu().numpy()
        return {'timestamps': self.timestamps, 'criterions':self.criterions,
                'acc': self.test_acc, 'F1': self.test_F1, 'losses': self.losses}

    def run_task(self, task_dic, shot):
        
        # Extract support and query
        y_s, y_q = task_dic['y_s'], task_dic['y_q']
        x_s, x_q = task_dic['x_s'], task_dic['x_q']

        # Transfer tensors to GPU if needed
        support = x_s.to(self.device)  # [ N * (K_s + K_q), d]
        query = x_q.to(self.device)  # [ N * (K_s + K_q), d]
        y_s = y_s.long().to(self.device)
        y_q = y_q.long().to(self.device)

        # Extract features
        support = support.to(self.device).float()
        query = query.to(self.device).float()

        # Perform normalizations required
        #scaler = MinMaxScaler(feature_range=(0, 1))
        #query, support = scaler(query, support)

        # Init basic prototypes
        self.init_weights(support=support, y_s=y_s, query=query, y_q=y_q)

        # Run adaptation
        self.run_adaptation(support=support, query=query, y_s=y_s, y_q=y_q, shot=shot)

class PADDLE(KM):
    def __init__(self, estimate_cov=False, diagonal=False, unique_covariance=False, reg=False):
        self.estimate_cov = estimate_cov
        self.diagonal = diagonal
        self.unique_covariance = unique_covariance
        self.reg = reg
        super().__init__()

    def p_update(self, query):
        """
        inputs:
            query : torch.Tensor of shape [n_task, q_shot, feature_dim]
        """
        
        q_shot = query.size(1)
        if self.diagonal == True:
            if self.unique_covariance == True:
                s = self.s.unsqueeze(1)
            else:
                s = self.s
            u = - 1 / 2 * self.get_logits(query, s).detach() + 1 / 2 * (torch.log(s).sum(2)).unsqueeze(1)
        
        else:
            if self.unique_covariance == True:
                s = self.s.unsqueeze(1)
                c = self.c.unsqueeze(1)
            else:
                s = self.s
                c = self.c
                
            w = self.weights
            
            if self.unique_covariance == True:
                s = self.s.unsqueeze(1)
                c = self.c.unsqueeze(1)
                u2 = w.unsqueeze(1) - query.unsqueeze(2)
                mat = c.unsqueeze(1).expand(u2.size(0), u2.size(1), u2.size(2), u2.size(3), u2.size(3))
                u = - 1 / 2 * (u2 * (torch.linalg.solve(mat, u2))).sum(3)
            else:
                s = self.s
                c = self.c
                u2 = w.unsqueeze(1) - query.unsqueeze(2)
                mat = c.unsqueeze(1).expand(u2.size(0), u2.size(1), u2.size(2), u2.size(3), u2.size(3))
                u = - 1 / 2 * (u2 * (torch.linalg.solve(mat, u2))).sum(3)
            
            # u2 = w.unsqueeze(1) - query.unsqueeze(2)
            # u = - 1 / 2 * (u2 * ((s.unsqueeze(1).matmul(u2.unsqueeze(4))).squeeze(4))).sum(3)
            u += - 1 / 2 * torch.logdet(c).unsqueeze(1)
            
        self.p = (u + self.alpha * self.A_adj(self.v, q_shot)).softmax(2)
        
    def v_update(self):
        """
        inputs:
        """
        self.v = torch.log(self.A(self.p) + 1e-6) + 1

    def weights_update(self, support, query, y_s_one_hot):
        """
        Corresponds to w_k updates
        inputs:
            support : torch.Tensor of shape [n_task, s_shot, feature_dim]
            query : torch.Tensor of shape [n_task, q_shot, feature_dim]
            y_s_one_hot : torch.Tensor of shape [n_task, s_shot, n_ways]


        updates :
            self.weights : torch.Tensor of shape [n_task, num_class, feature_dim]
        """
        
        num = (query.unsqueeze(2) * self.p.unsqueeze(3)).sum(1)
        num += (support.unsqueeze(2) * y_s_one_hot.unsqueeze(3)).sum(1)
        den  = self.p.sum(1) + y_s_one_hot.sum(1)
        self.weights = torch.div(num, den.unsqueeze(2)) 


    def s_update(self, support, query, y_s_one_hot):
        """
        inputs:
        """
        w = self.weights   # N x K x C

        if self.reg == False:
            if self.diagonal == True:
                d_s = (w.unsqueeze(1) - support.unsqueeze(2)).square().mul(y_s_one_hot.unsqueeze(3)).sum(1) # N x K x C
                d_q = (w.unsqueeze(1) - query.unsqueeze(2)).square().mul(self.p.unsqueeze(3)).sum(1)
                d = (d_s + d_q)

                if self.unique_covariance == True:
                    self.s = ((self.p.sum(1) + y_s_one_hot.sum(1)).sum(1)).unsqueeze(1) / d.sum(1)
                    self.c = 1 / (self.s.clamp(min=1e-6))
                else:
                    self.s = (self.p.sum(1) + y_s_one_hot.sum(1)).unsqueeze(2) / d
                    self.c = 1 / (self.s.clamp(min=1e-6))
            else :
                if self.unique_covariance == True:
                    u = (w.unsqueeze(1) - support.unsqueeze(2)) 
                    C = (y_s_one_hot.unsqueeze(3).unsqueeze(3) * (u.unsqueeze(4).matmul(u.unsqueeze(3)))).sum(1).sum(1)
                    u = (w.unsqueeze(1) - query.unsqueeze(2)) 
                    C += (self.p.unsqueeze(3).unsqueeze(3) * (u.unsqueeze(4).matmul(u.unsqueeze(3)))).sum(1).sum(1)
                    C = C / ((self.p.sum(1) + y_s_one_hot.sum(1)).sum(1)).unsqueeze(1).unsqueeze(1).clamp(min=1e-6) 
                    self.c = C
                    self.s = torch.inverse(C) 

                else:
                    u = (w.unsqueeze(1) - support.unsqueeze(2)) 
                    C = (y_s_one_hot.unsqueeze(3).unsqueeze(3) * (u.unsqueeze(4).matmul(u.unsqueeze(3)))).sum(1)
                    u = (w.unsqueeze(1) - query.unsqueeze(2)) 
                    C += (self.p.unsqueeze(3).unsqueeze(3) * (u.unsqueeze(4).matmul(u.unsqueeze(3)))).sum(1)
                    C = C / (self.p.sum(1) + y_s_one_hot.sum(1)).unsqueeze(2).unsqueeze(2).clamp(min=1e-6) 
                    self.c = C
                    self.s = torch.inverse(C) 
                    
        elif self.reg == True:
        
            if self.diagonal == True:
                d_s = (w.unsqueeze(1) - support.unsqueeze(2)).square().mul(y_s_one_hot.unsqueeze(3)).sum(1) # N x K x C
                d_q = (w.unsqueeze(1) - query.unsqueeze(2)).square().mul(self.p.unsqueeze(3)).sum(1)
                d = (d_s + d_q)
                eta = 0.0001

                if self.unique_covariance == True:
                    grad = d.sum(1) - (self.p.sum(1) + y_s_one_hot.sum(1)).sum(1).unsqueeze(1) * self.c
                    grad = grad / 2 
                    self.s = self.proj_box(self.s - eta * grad)
                    self.c = 1 / self.s.clamp(min=1e-6) 

                else:
                    grad = d - (self.p.sum(1) + y_s_one_hot.sum(1)).unsqueeze(2) * self.c
                    grad = grad / 2 
                    self.s = self.proj_box(self.s - eta * grad)
                    self.c = 1 / self.s.clamp(min=1e-6) 
            else :    
                eta = 0.001
                if self.unique_covariance == True:
                    u = (w.unsqueeze(1) - support.unsqueeze(2)) 
                    grad = (y_s_one_hot.unsqueeze(3).unsqueeze(3) * (u.unsqueeze(4).matmul(u.unsqueeze(3)))).sum(1).sum(1)
                    u = (w.unsqueeze(1) - query.unsqueeze(2)) 
                    grad += (self.p.unsqueeze(3).unsqueeze(3) * (u.unsqueeze(4).matmul(u.unsqueeze(3)))).sum(1).sum(1)
                    grad += - (self.p.sum(1) + y_s_one_hot.sum(1)).sum(1).unsqueeze(1).unsqueeze(1) * self.c
                    grad = grad / 2 
                    self.s = self.proj_trace(self.s - eta * grad)
                    # self.c = torch.inverse(self.s) 

                else:
                    u = (w.unsqueeze(1) - support.unsqueeze(2)) 
                    grad = (y_s_one_hot.unsqueeze(3).unsqueeze(3) * (u.unsqueeze(4).matmul(u.unsqueeze(3)))).sum(1)
                    u = (w.unsqueeze(1) - query.unsqueeze(2)) 
                    grad += (self.p.unsqueeze(3).unsqueeze(3) * (u.unsqueeze(4).matmul(u.unsqueeze(3)))).sum(1)
                    grad += - (self.p.sum(1) + y_s_one_hot.sum(1)).unsqueeze(2).unsqueeze(2) * self.c
                    grad = grad / 2 
                    self.s = self.proj_trace(self.s - eta * grad)
                    # self.c = torch.inverse(self.s) 
            

        
    def run_adaptation(self, support, query, y_s, y_q, shot):
        """
        Corresponds to the TIM-ADM inference
        inputs:
            support : torch.Tensor of shape [n_task, s_shot, feature_dim]
            query : torch.Tensor of shape [n_task, q_shot, feature_dim]
            y_s : torch.Tensor of shape [n_task, s_shot]
            y_q : torch.Tensor of shape [n_task, q_shot]


        updates :
            self.weights : torch.Tensor of shape [n_task, num_class, feature_dim]
        """
        
        y_s_one_hot = get_one_hot(y_s).float()
        feature_dim = support.size(2)
        n_task, n_ways = y_s_one_hot.size(0), y_s_one_hot.size(2)
        self.v = torch.zeros(n_task, n_ways).to(self.device)
        if self.diagonal == True:
            if self.unique_covariance == True:
                self.s = torch.ones(n_task, feature_dim).to(self.device)
                self.c = deepcopy(self.s).to(self.device)
            else:
                self.s = torch.ones(n_task, n_ways, feature_dim).to(self.device)
                self.c = deepcopy(self.s).to(self.device)
        else:
            if self.unique_covariance == True:
                self.s = torch.eye(feature_dim).repeat(n_task, 1, 1).to(self.device) 
                self.c = deepcopy(self.s).to(self.device)
            else:
                self.s = torch.eye(feature_dim).repeat(n_task, n_ways, 1, 1).to(self.device) 
                self.c = deepcopy(self.s).to(self.device)
            
        # self.p = get_one_hot(y_q).float()
        for i in tqdm(range(self.iter)):
            weights_old = deepcopy(self.weights.detach())
            t0 = time.time()
            self.p_update(query)
            self.v_update()
            self.weights_update(support, query, y_s_one_hot)
            if self.estimate_cov == True:
                self.s_update(support, query, y_s_one_hot)
            
            weight_diff = (weights_old - self.weights).norm(dim=-1).mean(-1)
            criterions = weight_diff
            t1 = time.time()
            self.record_convergence(new_time=t1-t0, criterions=criterions)
            
            #if math.isnan(self.s.sum()):
             #   break
        #print(self.p.sum())
            # print("p", self.p)
        # print("w", self.weights)
            #print("s", self.s)
        # print(self.p)
        self.record_info(y_q=y_q)
        
class MinMaxScaler(object):
    """MinMax Scaler

    Transforms each channel to the range [a, b].

    Parameters
    ----------
    feature_range : tuple
        Desired range of transformed data.
    """

    def __init__(self, **kwargs):
        self.__dict__.update(kwargs)

    def __call__(self, query, support):
        """Fit features

        Parameters
        ----------
        stacked_features : tuple, list
            List of stacked features.

        Returns
        -------
        tensor 
            A tensor with scaled features using requested preprocessor.
        """

        dist = (query.max(dim=1, keepdim=True)[0] - query.min(dim=1, keepdim=True)[0])
        dist[dist==0.] = 1.
        scale = 1.0 /  dist
        ratio = query.min(dim=1, keepdim=True)[0]
        query.mul_(scale).sub_(ratio)
        support.mul_(scale).sub_(ratio)
        return query, suppor
```

```python
from numpy.random import multivariate_normal
import matplotlib.pyplot as plt
np.random.seed(0)
qshot = 75
shot = 20
n_task = 2
n_ways = 5
d = 512
sshot = n_ways * shot
eta = 0.0
zeta = 0.1
prob = 1.0

task_dic = {"x_q":[], "x_s":[], "y_q":[], "y_s":[]}
list_cov = []
list_mean = []

for k in range(n_ways):
    m = np.zeros(d) 
    m[k] = 1
    list_mean.append(m)    
diag = zeta + (np.random.random(d) * (2 * zeta * eta) - eta * zeta)
u = np.random.random(d)
cov = np.diag(diag * (u <= prob) + 1 * (u > prob))
#print(1 / np.diag(cov))
list_cov.append(cov)
    
for task in range(n_task):
    for k in range(n_ways):
        task_dic["x_q"].append(multivariate_normal(list_mean[k], list_cov[0], size=(qshot//n_ways)))
        task_dic["x_s"].append(multivariate_normal(list_mean[k], list_cov[0], size=(shot)))
        task_dic["y_q"].append(k * np.ones(qshot//n_ways))
        task_dic["y_s"].append(k * np.ones(shot))
```
